```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.constants
from sympy.physics.wigner import wigner_6j
```

```python
""" 
constants
"""
pi = np.pi
c = sp.constants.c
epsilon_0 = sp.constants.epsilon_0
hbar = sp.constants.hbar
h = sp.constants.h
```

From: 

Fam Le Kien, P. Schneeweiss, and A. Rauschenbeutel. Dynamical polarizability of atoms in arbitrary light fields: general theory and
application to cesium}. (2012) https://arxiv.org/pdf/1211.2673.pdf

Without hyperfine splitting:

1. Scalar polarizability:
$$\alpha^s_{nJ} = \frac{1}{\sqrt{3(2J+1)}} \alpha^{(0)}_{nJ}$$

2. Vector polarizability:
$$\alpha^v_{nJ} = -\sqrt{\frac{2J}{(J+1)(2J+1)}} \alpha^{(1)}_{nJ}$$

3. Tensor polarizability:
$$\alpha^t_{nJ} = -\sqrt{\frac{2J(2J-1)}{3(J+1)(2J+1)(2J+3)}} \alpha^{(2)}_{nJ}$$

$\alpha^s_{nJ}$ and $\alpha^t_{nJ}$ for particular $nJ$ state from Ryuta Yamamoto et al 2016 New J. Phys. 18 023016 https://iopscience.iop.org/article/10.1088/1367-2630/18/2/023016/pdf


With hyperfine splitting:

1. Scalar polarizability:
$$\alpha^s_{nJF} = \alpha^s_{nJ}$$

2. Vector polarizability:
$$\alpha^v_{nJF} = (-1)^{J+I+F+1}\sqrt{\frac{(J+1)(2J+1)}{2J}}\sqrt{\frac{2F(2F+1)}{F+1}}
\begin{Bmatrix}
F & 1 & F\\
J & I & J
\end{Bmatrix} 
\alpha^{v}_{nJ}$$

3. Tensor polarizability:
$$\alpha^t_{nJF} = (-1)^{J+F+I}\sqrt{\frac{3(J+1)(2J+1)(2J+3)}{2J(2J-1)}}\sqrt{\frac{2F(2F-1)(2F+1)}{3(F+1)(2F+3)}}
\begin{Bmatrix}
F & 2 & F\\
J & I & J
\end{Bmatrix} 
\alpha^{t}_{nJ}$$


Light shift:
\begin{equation}
V = -\frac{1}{4}|E|^2\{
\alpha^s_{nJF}
-i\alpha^v_{nJF}\frac{[\mathbf{u^*} \times \mathbf{u}] \cdot \mathbf{F}}{2F}
+ \alpha^t_{nJF}\frac{3[(\mathbf{u^*}\cdot\mathbf{F})(\mathbf{u}\cdot\mathbf{F}) + (\mathbf{u}\cdot\mathbf{F})(\mathbf{u^*}\cdot\mathbf{F})]-2\mathbf{F}^2}{2F(2F-1)}
\}
\end{equation}

with the E-filed being $E\mathbf{u}$. 


For linearly polarized light, light shift as the function of the angle $\theta$ between the light polarization and the quantization axis:
\begin{equation}
V = -\frac{1}{4}I ( \alpha^s_{nJF} - \alpha^t_{nJF}\frac{1-3\cos^2{\theta}}{2}(3m_F^2 - 2) )
\end{equation}

```python
class state(object):
    def __init__(self, J, I, F, m):
        self.J = J
        self.I = I
        self.F = F
        self.m = m
        
class polarizability_measured(object):
    def __init__(self, s, v, t):
        self.s = s
        self.v = v
        self.t = t
```

```python
""" 
polarizability expressions
i - state; a - measured polarizability without hyperfine splitting
"""

def alpha_scalar(i, a):
    return (
            a.s
            )

def alpha_vector(i, a):
    if i.J:
        return (
                (-1)**float(i.J + i.I + i.F + 1)
                * np.sqrt((i.J + 1)*(2*i.J+1) / (2*i.J))
                * np.sqrt(2*i.F * (2*i.F + 1) / (i.F + 1))
                * wigner_6j(i.F, 1, i.F, i.J, i.I, i.J)
                * a.v
                )
    else: return 0

def alpha_tensor(i, a):
    if i.J and (2*i.J-1):
        return (
                (-1)**float(i.J + i.F + i.I)
                * np.sqrt(3 * (i.J + 1) * (2*i.J + 1) * (2*i.J + 3) / (2*i.J) / (2*i.J -1) ) 
                * np.sqrt(2*i.F * (2*i.F -1) * (2*i.F + 1) / 3. / (i.F + 1) / (2*i.F + 3))
                * wigner_6j(i.F, 2, i.F, i.J, i.I, i.J)
                * a.t
                )
    else: return 0
```

```python
"""
Light shifts expressions
"""

def angle_shift(i, intensity, a):
    return lambda theta: ( - 1. / 4. * intensity 
                         *( alpha_scalar(i, a) - alpha_tensor(i, a) *
                         (1. - 3.*np.cos(theta)**2) / 2. * (3.*i.m**2 - 2.)
                         ))
```

```python
# plot range [pi]
l_range = [0, np.pi]

# divide plot range into n points
num_points = 1000

# lattice polatization 
# E_x, E_y, E_z
# can use imaginary numbers for elliptical polarizations
# e.g. [1j, 1, 0]
#lattice_field = np.array([0,0,1])

intensity = 1 #scalar, intensity Wcm-2

fig, ax = plt.subplots(figsize=(10,6))
L = np.linspace(l_range[0], l_range[1], num_points)

#1S0
pm_1S0 = polarizability_measured(s=37.9, v=0, t=0)

state_ground = state(J=0, I=1/2., F=1/2., m=1/2.)
shift_ground = angle_shift(state_ground, intensity, pm_1S0)(L) #Hz
ax.plot(L/np.pi, shift_ground, color='tab:red', linestyle='-', label='$Yb 171, Yb 174, {}^1S_0$')


#3P1
pm_3P1 = polarizability_measured(s=22.4, v=0, t=-7.6)

state_32_32 = state(J=1, I=1/2., F=3/2., m=3/2.)
shift_32_32 = angle_shift(state_32_32, intensity, pm_3P1)(L) #Hz
ax.plot(L/np.pi, shift_32_32, color='tab:cyan', linestyle='-', label='$Yb 171, {}^3P_1, F=3/2, m=3/2$')

state_32_12 = state(J=1, I=1/2., F=3/2., m=1/2.)
shift_32_12 = angle_shift(state_32_12, intensity, pm_3P1)(L) #Hz
ax.plot(L/np.pi, shift_32_12, color='tab:orange', linestyle='-', label='$Yb 171, {}^3P_1, F=3/2, m=1/2$')

state_12_12 = state(J=1, I=1/2., F=1/2., m=1/2.)
shift_12_12 = angle_shift(state_12_12, intensity, pm_3P1)(L) #Hz
ax.plot(L/np.pi, shift_12_12, color='tab:purple', linestyle='-', label='$Yb 171, {}^3P_1, F=1/2, m=1/2$')


#3P1 (Yb174 F=J=1, mF=0)
shift_Yb174 = ( - 1. / 4. * intensity 
                         *( 22.4 - (-7.6) *
                         (1. - 3.*np.cos(L)**2) / 2. * (3.*0.**2 - 2.)
                         )) #Hz
ax.plot(L/np.pi, shift_Yb174, color='tab:green', linestyle='-', label='$Yb174, {}^3P_1, m_J=0$')



ax.set_ylabel('Energy shift (Hz per Wcm-2)')
ax.set_xlabel('Angle ($\pi$)')
#ax.set_ylim(-4, 4)
#ax.set_xlim(400, 600)

ax.legend(bbox_to_anchor=[1., 0.5], loc='center left')

plt.tight_layout()
plt.savefig('1S0_3P1_linear.png', format='png', dpi=300)
plt.show()

print(float(min(shift_32_12)-shift_ground[0]))
print(float(min(shift_12_12)-shift_ground[0]))

idx = np.argwhere(np.diff(np.sign(shift_ground - shift_32_32))).flatten()
print(L[idx]/np.pi*180)
```
