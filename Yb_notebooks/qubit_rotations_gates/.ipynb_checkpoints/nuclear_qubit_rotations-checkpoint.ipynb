For large detuning and small spontaneous emission, the Rabi frequency of a Raman transition is given by

$$
\Omega_R = \frac{1}{2 \hbar^2} \sum_{e} \frac{\bra{f} \mathbf{d}\cdot\mathbf{E} \ket{e} \bra{e} \mathbf{d}\cdot\mathbf{E} \ket{i}}{\Delta_e}
$$

where $\ket{i}$ is the initial state, $\ket{f}$ is the final state, the sum runs over excited state $\ket{e}$ and $\Delta_e$ is the detuning of both Raman beams from the state $\ket{e}$. Individual hyperfine matrix elements can be decomposed into so many angular momentum pieces,

$$
\bra{F',m_F'} \mathbf{d} \ket{F,m_F} = (-1)^{F+J'+I+1} (2J'+1)\sqrt{\frac{2F+1}{2J+1}} \braket{F',m_F'}{F,m_F,1,q} \ws{J'}{J}{1}{F}{F'}{I} \left( \Gamma_{JJ'} \frac{3\pi \epsilon_0 \hbar c^3}{\omega_0^3} \right)^{1/2}
$$

and summed over.

For $^{171}$Yb Raman transitions for single qubit rotations in the ground state through $^3P_1$, $J=0$, $F=1/2$, $m_F=\pm1/2$, $J'=1$, $F'=1/2,3/2$, $m_F'=\pm1/2, \pm3/2$, and the above hyperfine dipole matrix elements are,

$$
\bra{F',m_F'} \mathbf{d} \ket{F,m_F} = -3\sqrt{2} \braket{F',m_F'}{1/2,m_F,1,q} \ws{1}{0}{1}{\frac{1}{2}}{F'}{\frac{1}{2}} \left( \Gamma_0 \frac{3\pi \epsilon_0 \hbar c^3}{\omega_0^3} \right)^{1/2}
$$

And the Raman Rabi freqeuncy is, assuming the same intensity of both beams $I_0$, with polarizations $q1$ and $q2$, and transition from $m_{F1}$ to $m_{F2}$ in the ground state,

$$
\Omega_R (m_{F1},m_{F2},I_0,B) =  \frac{I_0}{\epsilon_0 c \hbar^2} \sum_{F',mF'} \frac{\bra{F=1/2, m_{F2}} d_{q2} \ket{F',m_F'} \bra{F', m_F'} d_{q1} \ket{F=1/2, m_{F1}}}{\Delta_{F',m_F'}(B)}
$$

```python
sc.hbar*A0/muB
```

Here $\Delta_{F',m_F'}(B)$ is a function of the magnetic field, for field strengths $B\ll h \Delta_{HF}/(g_F \mu_B) \sim$ 700 G, for the hyperfine splitting $\Delta_{HF}\sim$ 6 GHz between $F'=1/2$ and $F'=3/2$. This assumes that Raman beams are still tuned relative to any shifts in the hyperfine ground states (they have the same detuning from all excited states for any field strength B). Then

$$
\Delta_{F',m_F'}(B) = \Delta_{0} - \left(F'-\frac{1}{2}\right)\Delta_{HF} + g_F \mu_B B m_F
$$

where $\Delta_0$ is the detuning relative to $F' = 1/2$ at $B=0$ and

$$
g_F = \frac{F(F+1) + J(J+1) - I(I+1)}{2F(F+1)} g_J \\
g_J = \frac{3}{2} + \frac{S(S+1) - L(L+1)}{2J(J+1)}
$$

For $^3P_1$ in $^{171}$Yb this is,

$$
g_F = \frac{1}{2} + \frac{15}{16 F(F+1)}
$$

The scattering rate is approximately $R_{sc} = \Gamma_0 \rho_{ee}$, which can be roughly estimated by

$$
R_{sc} = \frac{\Gamma_0 \left(\Omega_1^2 + \Omega_2^2\right)}{8\Delta^2}
$$

```python
# function for computing hyperfine matrix elements for Raman out of ground state 171Yb through 3P1
def hypefine_matrix_elem(mF, Fp, mFp, q, omega=omega0, Gamma=Gamma0):
    angular_part = -3*np.sqrt(2) * ( wigner.clebsch_gordan(1/2,1,Fp,mF,q,mFp) 
                                          * wigner.wigner_6j(1,0,1,1/2,Fp,1/2) )
    reduced_matrix_elem = np.sqrt( Gamma*3*np.pi*sc.epsilon_0*sc.hbar*(sc.c**3)/(omega**3) )
    return float(angular_part * reduced_matrix_elem)

hypefine_matrix_elem_v = np.vectorize(hypefine_matrix_elem)

def gF(F):
    return (1/2) + (15/(16*F*(F+1)))

def deltaHF(delta0, Fp, mFp, B, A):
    return 2*np.pi*delta0 - 2*np.pi*(Fp-0.5)*A - gF(Fp)*muB*B*mFp/sc.hbar

def raman_rabi(mF1, mF2, I0, q1, q2, B, delta0, omega=omega0, Gamma=Gamma0, A=A0):
    Ep = I0/(sc.epsilon_0*sc.c*(sc.hbar**2))
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    hmes = 0
    for FmF in FmFs:
        hmes = hmes + (hypefine_matrix_elem_v(mF2, FmF[0], FmF[1], q2, omega, Gamma)
                      * hypefine_matrix_elem(mF1, FmF[0], FmF[1], q1, omega, Gamma)
                    / deltaHF(delta0, FmF[0], FmF[1], B, A))
    return np.abs(Ep*hmes)/(2*pi)

#approximate scattering rate assuming large detuning, no cross coupling/coherences
def Rsc(mF1, mF2, I0, q1, q2, B, delta0, omega=omega0, Gamma=Gamma0, A=A0):
    Ep = np.sqrt(I0/(sc.epsilon_0*sc.c))
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    hmes1 = 0
    hmes2 = 0
    for FmF in FmFs:
        hmes1 = hmes1 + hypefine_matrix_elem_v(mF1, FmF[0], FmF[1], q1, omega, Gamma)/deltaHF(delta0, FmF[0], FmF[1], B, A)
        hmes2 = hmes2 + hypefine_matrix_elem_v(mF2, FmF[0], FmF[1], q2, omega, Gamma)/deltaHF(delta0, FmF[0], FmF[1], B, A)
    return (Ep**2)*Gamma*(hmes1**2 + hmes2**2)/(8*(sc.hbar**2))
```

## clock state  3S1 rotations

For $^{171}$Yb Raman transitions for single qubit rotations in the clock state through $^3S_1$, $J=0$, $F=1/2$, $m_F=\pm1/2$, $J'=1$, $F'=1/2,3/2$, $m_F'=\pm1/2, \pm3/2$, and the above hyperfine dipole matrix elements have the same form as above, but the natural linewidth is larger, $\Gamma_{3S1}=9.6$ MHz.

$$
\bra{F',m_F'} \mathbf{d} \ket{F,m_F} = -3\sqrt{2} \braket{F',m_F'}{1/2,m_F,1,q} \ws{1}{0}{1}{\frac{1}{2}}{F'}{\frac{1}{2}} \left( \Gamma_{3S1} \frac{3\pi \epsilon_0 \hbar c^3}{\omega_{3S1}^3} \right)^{1/2}
$$

And the Raman Rabi freqeuncy is, assuming beam intensities $I_1$ and $I_2$, with polarizations $q1$ and $q2$, and transition from $m_{F1}$ to $m_{F2}$ in the ground state,

$$
\Omega_R (m_{F1},m_{F2},I_1,I_2,B) =  \frac{\sqrt{I_1 I_2}}{\epsilon_0 c \hbar^2} \sum_{F',mF'} \frac{\bra{F=1/2, m_{F2}} d_{q2} \ket{F',m_F'} \bra{F', m_F'} d_{q1} \ket{F=1/2, m_{F1}}}{\Delta_{F',m_F'}(B)}
$$

# with cross coupling

Generalize above for beams with multiple components of polarization and include cross coupling of laser beams to the ground states. With cross coupling, but still RWA for the optical frequency, the atom-field coupling looks like,

$$
H_{AF} = \frac{1}{2}\left\{ \Omega_1 \left( \sigma_1 e^{i \omega_1 t} +\sigma^\dagger_1 e^{-i \omega_1 t} \right)
+ \Omega_2 \left( \sigma_2 e^{i \omega_2 t} +\sigma^\dagger_2 e^{-i \omega_2 t} \right)
+ \Omega_1' \left( \sigma_2 e^{i \omega_2 t} +\sigma^\dagger_2 e^{-i \omega_2 t} \right)
+ \Omega_2' \left( \sigma_1 e^{i \omega_1 t} +\sigma^\dagger_1 e^{-i \omega_1 t} \right)
\right\}
$$

where $\sigma_\alpha = \ket{\alpha}\bra{e}$ and the last two terms represent the cross coupling. The Rabi rates $\Omega_\alpha'$ for cross coupling are distinguished from the others to allow for the possibility of different polarization coupling in the two different directions. In the rotating frame given by the transformation,

$$
U=exp\left\{-i \left( \omega_1 t \ket{g_1}\bra{g_1} + \omega_2 t \ket{g_2}\bra{g_2} \right)  \right\}
$$

the Hamiltonian becomes, $\tilde{H} = U H U^\dagger +i(\partial_t U)U^\dagger$, absorbing the second term into the $\tilde{H}_{A}$, it becomes,

$$
\tilde{H}_{A} = \Delta_1 \ket{g_1}\bra{g_1} + \Delta_2 \ket{g_2}\bra{g_2}
$$

where $\Delta_{\alpha} = \omega_\alpha - \omega_{0\alpha}$. And the atom-field interaction becomes,

$$
\tilde{H}_{AF} = \frac{1}{2}\left\{ 
\left(\Omega_1 +\Omega_2' e^{i\Delta_g t} \right) \sigma_1
+\left(\Omega_1' e^{-i\Delta_g t} +\Omega_2  \right) \sigma_2
+\left(\Omega_1 +\Omega_2' e^{-i\Delta_g t} \right) \sigma_1^\dagger
+\left(\Omega_1' e^{i\Delta_g t} +\Omega_2  \right) \sigma_2 ^\dagger
\right\}
$$

Ignoring spontaneous emission for now, boosting by $-\Delta=(\Delta_1+\Delta_2)/2$ as in Steck, the components of the state have time dependence,
$$
\partial_t \psi_e = \frac{-i}{2} \left\{ 
\left( \Omega_1 +\Omega_2' e^{i\Delta_g t} \right) \psi_{1}
+\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right) \psi_{2}
-\Delta \psi_e
\right\}
$$

$$
\partial_t \psi_1 = -i \left\{ 
\frac{1}{2}\left( \Omega_1 +\Omega_2'e^{-i\Delta_g t} \right) \psi_e
+(\Delta_1-\Delta) \psi_1
\right\}
$$

$$
\partial_t \psi_2 = -i \left\{ 
\frac{1}{2}\left( \Omega_1'e^{i\Delta_g t} +\Omega_2 \right) \psi_e
+(\Delta_2-\Delta) \psi_2
\right\}
$$

For $\Gamma \ll \Delta$, take the approximation $\partial_t \psi_e = 0$, and this gives

$$
\psi_e = \frac{1}{2 \Delta} \left\{ 
\left( \Omega_1 +\Omega_2' e^{i\Delta_g t} \right) \psi_{1}
+\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right) \psi_{2} \right\}
$$

and plugging this into the equations for $\psi_1$ and $\psi_2$,

$$
\partial_t \psi_1 = -i \left\{ 
\frac{1}{4\Delta}\left( \Omega_1 +\Omega_2'e^{-i\Delta_g t} \right)\left\{ 
\left( \Omega_1 +\Omega_2' e^{i\Delta_g t} \right) \psi_{1}
+\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right) \psi_{2} \right\}
+(\Delta_1-\Delta) \psi_1
\right\}
$$

$$
\partial_t \psi_1 = -i \left\{ \left[ (\Delta_1-\Delta) + \frac{1}{4\Delta} \left( \Omega_1^2 \Omega_2'^2 +2\Omega_1 \Omega_2' cos(\Delta_g t) \right) \right] \psi_1 + \frac{1}{2} \frac{\left(\Omega_1 +\Omega_2'e^{-i\Delta_g t} \right)\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right)}{2\Delta}\psi_2
\right\}
$$

$$
\partial_t \psi_2 = -i \left\{ 
\frac{1}{4\Delta}\left( \Omega_1'e^{i\Delta_g t} +\Omega_2 \right)\left\{ 
\left( \Omega_1 +\Omega_2' e^{i\Delta_g t} \right) \psi_{1}
+\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right) \psi_{2} \right\}
+(\Delta_2-\Delta) \psi_1
\right\}
$$

$$
\partial_t \psi_2 = -i \left\{ \left[ (\Delta_2-\Delta) + \frac{1}{4\Delta} \left( \Omega_1'^2 \Omega_2^2 +2\Omega_1' \Omega_2 cos(\Delta_g t) \right) \right] \psi_2 + \frac{1}{2} \frac{\left(\Omega_1'e^{i\Delta_g t} +\Omega_2 \right)\left( \Omega_1 +\Omega_2'e^{i\Delta_g t} \right)}{2\Delta}\psi_1
\right\}
$$

Define the stark shifts and Raman Rabi frequency

$$
\omega_{AC1} = \frac{1}{4\Delta} \left( \Omega_1^2 + \Omega_2'^2 +2\Omega_1 \Omega_2' cos(\Delta_g t) \right) \\
\omega_{AC2} = \frac{1}{4\Delta} \left( \Omega_1'^2 + \Omega_2^2 +2\Omega_1' \Omega_2 cos(\Delta_g t) \right)
$$

$$
\Omega_R(t) = \frac{\left(\Omega_1 +\Omega_2'e^{-i\Delta_g t} \right)\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right)}{2\Delta}
$$

This gives

$$
\partial_t \psi_1 = -i \left\{ \left[ (\Delta_1-\Delta) + \omega_{AC1} \right] \psi_1 + \frac{1}{2} \Omega_R\psi_2
\right\}
$$

$$
\partial_t \psi_2 = -i \left\{ \left[ (\Delta_2-\Delta) + \omega_{AC2} \right] \psi_2 + \frac{1}{2} \Omega_R^*\psi_1
\right\}
$$

and an effective two-level Hamiltonian

$$
H_{eff} = \begin{pmatrix}
(\Delta_2-\Delta)+\omega_{AC2} & \frac{1}{2} \Omega_R^* \\
\frac{1}{2} \Omega_R & (\Delta_1-\Delta)+\omega_{AC1}
\end{pmatrix}
$$

Note that here if there is a difference between the stark shifts, which I see below caused by cross coupling, then the effective Hamiltonian looks like

$$
\tilde{H}_{eff} = \begin{pmatrix}
-\Delta_{raman} & \frac{1}{2} \Omega_R^* \\
\frac{1}{2} \Omega_R & 0
\end{pmatrix}
$$

when rotating the frame with $U=e^{-i ((\Delta_1-\Delta)+\omega_{AC1})t}$. This means there is effectively a detuning from the Raman transition given by $\Delta_{raman} = \Delta_1-\Delta_2 + \omega_{AC1}-\omega_{AC2}$ and **the fidelity will be limited by our ability to cancel the difference in stark shifts**.

Including the dependence on polarization $\mathbf{E} = E \hat{\epsilon}_L= E(e_+\hat{\epsilon}_+ +e_-\hat{\epsilon}_- + e_0\hat{\epsilon}_0)$ and hyperfine structure of excited levels, the bare Rabi frequencies are

$$
\Omega_{\alpha,F',m_F'} = \left(\frac{2 I_0}{\epsilon_0 c}\right)^{1/2}\frac{\bra{F', m_F'} e_+ d_{+} + e_- d_{-} + e_0 d_{0}  \ket{F=1/2, m_{\alpha}}}{\hbar}
$$

where again

$$
\bra{F',m_F'} \mathbf{d} \ket{F,m_F} = -3\sqrt{2} \braket{F',m_F'}{1/2,m_F,1,q} \ws{1}{0}{1}{\frac{1}{2}}{F'}{\frac{1}{2}} \left( \Gamma_0 \frac{3\pi \epsilon_0 \hbar c^3}{\omega_0^3} \right)^{1/2}
$$ 

$$
\omega_{AC1} = \sum_{F',m_F'} \frac{\left( \Omega_1^2 + \Omega_2'^2 +2\Omega_1 \Omega_2' cos(\Delta_g t) \right)}{4\Delta} \\
\omega_{AC2} = \sum_{F',m_F'} \frac{\left( \Omega_1'^2 + \Omega_2^2 +2\Omega_1' \Omega_2 cos(\Delta_g t) \right)}{4\Delta} \\
\Omega_R(t) = \sum_{F',m_F'}\frac{\left(\Omega_1 +\Omega_2'e^{-i\Delta_g t} \right)\left( \Omega_1'e^{-i\Delta_g t} +\Omega_2 \right)}{2\Delta}
$$

For this part, everything is in angular frequencies, including detunings and frequency splittings.

```python
# a function for computing the bare rabi frequencies above, pol is a vector with components [e+, e-, e0] defined above

def deltaFmF(delta0, Fp, mFp, B):
    return delta0 - (Fp-0.5)*2*np.pi*A0 - gF(Fp)*muB*B*mFp/sc.hbar

def Omega_bare(I0,mF,Fp,mFp,pol):
    return np.sqrt(2*I0/(sc.epsilon_0*sc.c))*(pol[0]*hypefine_matrix_elem(mF, Fp, mFp, 1) 
                                       + pol[1]*hypefine_matrix_elem(mF, Fp, mFp, -1) 
                                       + pol[2]*hypefine_matrix_elem(mF, Fp, mFp, 0))/sc.hbar

def omega_AC(I0,mF,pol1,pol2,B,t):
    dg = B*muI/sc.hbar
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    omega_AC = 0
    for FmF in FmFs:
        omega_AC = omega_AC + ( Omega_bare(I0,mF,FmF[0],FmF[1],pol1)**2 
                                 + Omega_bare(I0,mF,FmF[0],FmF[1],pol2)**2
                               + 2*Omega_bare(I0,mF,FmF[0],FmF[1],pol1)*Omega_bare(I0,mF,FmF[0],FmF[1],pol2)*np.cos(dg*t) 
                              )/(4*deltaFmF(delta0, FmF[0], FmF[1], B))
    return omega_AC

def Omega_Rabi(I0,mF1,mF2,pol1,pol2,delta0,B,t):
    dg = B*muI/sc.hbar
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    Omega_R = 0
    for FmF in FmFs:
        Omega_R = Omega_R + ( (Omega_bare(I0,mF1,FmF[0],FmF[1],pol1) + Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)*np.exp(-1j*dg*t))
                             *(Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)*np.exp(-1j*dg*t) + Omega_bare(I0,mF2,FmF[0],FmF[1],pol2))
                             /(2*deltaHF(delta0, FmF[0], FmF[1], B)) )
    return Omega_R
```

```python
def raman_eff(psi, t, oAC1, oAC2, o1, o2, o1p, o2p, dg, d):
    c1, c2 = psi
    return np.array([-1j*(oAC1*c1 + (1/(4*d))*(o1+o2p*np.exp(-1j*dg*t))*(o1p*np.exp(-1j*dg*t)+o2)*c2),
           -1j*(oAC2*c2 + (1/(4*d))*(o1+o2p*np.exp(1j*dg*t))*(o1p*np.exp(1j*dg*t)+o2)*c1)])

def raman_jac(psi, t, oAC1, oAC2, o1, o2, o1p, o2p, dg, d):
    c1, c2 = psi
    return np.array([[-1j*oAC1, -1j*(1/(4*d))*(o1+o2p*np.exp(-1j*dg*t))*(o1p*np.exp(-1j*dg*t)+o2)],
            [-1j*(1/(4*d))*(o1+o2p*np.exp(1j*dg*t))*(o1p*np.exp(1j*dg*t)+o2)], -1j*oAC2])
```

## calculating with zeeman and hyperfine

```python
def raman_eff_full(psi, t, Omega1s, Omega1ps, Omega2s, Omega2ps, deltaFmFs, delta0, delta01, delta02):
    c1, c2 = psi
    dg = B*muI/sc.hbar
    Omega_R = np.sum( (Omega1s + Omega2ps*np.exp(-1j*dg*t))*(Omega1ps*np.exp(-1j*dg*t) + Omega2s)/(2*deltaFmFs) )
    omega_AC1 = np.sum( ( Omega1s**2 + Omega2ps**2 + 2*Omega1s*Omega2ps*np.cos(dg*t) )/(4*deltaFmFs) )
    omega_AC2 = np.sum( ( Omega1ps**2 + Omega2s**2 + 2*Omega1ps*Omega2s*np.cos(dg*t) )/(4*deltaFmFs) )
    return -1j*np.array([(delta01 - delta0 + omega_AC1)*c1 + 0.5*np.conj(Omega_R)*c2,
           0.5*Omega_R*c1 + (delta02 - delta0 + omega_AC2)*c2])

def raman_jac_full(psi, t, Omega1s, Omega1ps, Omega2s, Omega2ps, deltaFmFs, delta0, delta01, delta02):
    c1, c2 = psi
    dg = B*muI/sc.hbar
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    Omega_R = np.sum( (Omega1s + Omega2ps*np.exp(-1j*dg*t))*(Omega1ps*np.exp(-1j*dg*t) + Omega2s)/(2*deltaFmFs) )
    omega_AC1 = np.sum( ( Omega1s**2 + Omega2ps**2 + 2*Omega1s*Omega2ps*np.cos(dg*t) )/(4*deltaFmFs) )
    omega_AC2 = np.sum( ( Omega1ps**2 + Omega2s**2 + 2*Omega1ps*Omega2s*np.cos(dg*t) )/(4*deltaFmFs) )
    return -1j*np.array([[delta01 - delta0 + omega_AC1, 0.5*np.conj(Omega_R)], 
                         [0.5*Omega_R, delta02 - delta0 + omega_AC2]])
```

```python
def raman_eff_full(psi, t, mF1, mF2, B, delta0, pol1, pol2):
    c1, c2 = psi
    dg = B*muI/sc.hbar
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    Omega_R = 0
    omega_AC1 = 0
    omega_AC2 = 0
    for FmF in FmFs:
        Omega_R = Omega_R + ( (Omega_bare(I0,mF1,FmF[0],FmF[1],pol1) + Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)*np.exp(-1j*dg*t))
                             *(Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)*np.exp(-1j*dg*t) + Omega_bare(I0,mF2,FmF[0],FmF[1],pol2))
                             /deltaFmF(delta0, FmF[0], FmF[1], B) )
        omega_AC1 = omega_AC1 + ( Omega_bare(I0,mF1,FmF[0],FmF[1],pol1)**2 
                                 + Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)**2
                               + 2*Omega_bare(I0,mF1,FmF[0],FmF[1],pol1)*Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)*np.cos(dg*t) 
                              )/(4*deltaFmF(delta0, FmF[0], FmF[1], B))
        omega_AC2 = omega_AC2 + ( Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)**2 
                                 + Omega_bare(I0,mF2,FmF[0],FmF[1],pol2)**2
                               + 2*Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)*Omega_bare(I0,mF2,FmF[0],FmF[1],pol2)*np.cos(dg*t) 
                              )/(4*deltaFmF(delta0, FmF[0], FmF[1], B))
    return np.array([-1j*(omega_AC1*c1 + np.conj(Omega_R)*c2),
           -1j*(Omega_R*c1 + omega_AC2*c2)])

def raman_jac_full(psi, t, mF1, mF2, B, delta0, pol1, pol2):
    c1, c2 = psi
    dg = B*muI/sc.hbar
    FmFs = np.array([[1/2,-1/2],[1/2,1/2],[3/2,-3/2],[3/2,-1/2],[3/2,1/2],[3/2,3/2]])
    Omega_R = 0
    omega_AC1 = 0
    omega_AC2 = 0
    for FmF in FmFs:
        Omega_R = Omega_R + ( (Omega_bare(I0,mF1,FmF[0],FmF[1],pol1) + Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)*np.exp(-1j*dg*t))
                             *(Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)*np.exp(-1j*dg*t) + Omega_bare(I0,mF2,FmF[0],FmF[1],pol2))
                             /deltaFmF(delta0, FmF[0], FmF[1], B) )
        omega_AC1 = omega_AC1 + ( Omega_bare(I0,mF1,FmF[0],FmF[1],pol1)**2 
                                 + Omega_bare(I0,mF1,FmF[0],FmF[1],pol2)**2
                               + 2*Omega_bare(I0,mF1,FmF[0],FmF[1],pol1)*Omega_bare(I0,mF,FmF[0],FmF[1],pol2)*np.cos(dg*t) 
                              )/(4*deltaFmF(delta0, FmF[0], FmF[1], B))
        omega_AC2 = omega_AC2 + ( Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)**2 
                                 + Omega_bare(I0,mF2,FmF[0],FmF[1],pol2)**2
                               + 2*Omega_bare(I0,mF2,FmF[0],FmF[1],pol1)*Omega_bare(I0,mF,FmF[0],FmF[1],pol2)*np.cos(dg*t) 
                              )/(4*deltaFmF(delta0, FmF[0], FmF[1], B))
    return -1j*np.array([[omega_AC1, np.conj(Omega_R)], [Omega_R, omega_AC2]])
```
