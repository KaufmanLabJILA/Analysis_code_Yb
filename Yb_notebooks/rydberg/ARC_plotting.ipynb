```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c6r = np.loadtxt("nc6r.txt", delimiter=',')
```

```python
colorsrc = plt.rcParams["axes.prop_cycle"].by_key()["color"]
```

```python
fig, ax = plt.subplots(nrows=2, figsize=[4,5],sharex=True)
ax[0].semilogy(c6r[0], c6r[1],'.',color=colorsrc[3], markersize=5)
ax[0].set_ylabel(r"$C_6$ (GHz $\mu$m$^6$)")
ax[1].plot(c6r[0],c6r[2],'.',color=colorsrc[0], markersize=5)
ax[1].set_ylabel(r"$r_b$ ($\mu$m)")
ax[1].set_xlabel(r"$n$")
# ax[0].tick_params(top=False,right=False,direction='out')
# plt.legend()
# plt.tight_layout()
plt.subplots_adjust(hspace=0.0)
plt.savefig('/Users/alec/JILA/NRC/C6.pdf')
```
