```python
import allantools
import random
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = [0 for i in range(10000)]
for i in range(10000):
    data[i] = random.gauss(0,1) + .1*i/10000
    
taus = np.linspace(1, 10000, 10000)
```

```python
plt.plot(taus,data)
```

```python
statistics.stdev(data)/np.sqrt(10000)
```

```python
(t2, ad, ade, adn) = allantools.oadev(data, rate=1, data_type="freq", taus=taus)

plt.errorbar(t2,ad, yerr=ade)
plt.xscale("log")
plt.yscale("log")
plt.xlabel('Time Cluster (sec)')
plt.ylabel('Allan Deviation')
```

```python
(t2, ad, ade, adn) = allantools.oadev(data[1:5000], rate=1, data_type="freq", taus=taus[1:5000])

plt.errorbar(t2,ad,yerr=ade)
plt.xscale("log")
plt.yscale("log")
plt.xlabel('Time Cluster (sec)')
plt.ylabel('Allan Deviation')
```
